```python
class QRS:
    def __init__(self, raw_data, sample_num): # raw_data: upper signal
        self.sec_step = 1/360 # 360 samples/sec
        volt_step = 10/(2**11) # 10mV into 11-bit
        self.sample_num = sample_num
        
        total_time = self.sec_step * self.sample_num
        
        data = raw_data * volt_step
        data -= data[0] # move y-axis (first sample == 0)
        self.data = data

        R = find_peaks(self.data, prominence = 1)
        
        R = R[0]# array (index of prominence peak)
        S = []
        Q = []
        
        # get idx of S and Q
        for i in range(len(R)):
            a = R[i]
            while self.data[a] - self.data[a-1] >= 0:
                a -= 1
                if a<1 : break
            Q.append(a)
            
            a = R[i]
            while self.data[a+1] - self.data[a] <= 0:
                a += 1
                if a+1 >= self.sample_num: break
            S.append(a)
            
        self.Q_idx = Q
        self.R_idx = R
        self.S_idx = S
    
    def plot(self): # plot all data
        plt.figure(figsize=(18,4))
        plt.plot(self.Q_idx, self.data[self.Q_idx], "xg")
        plt.plot(self.R_idx, self.data[self.R_idx], "ob")
        plt.plot(self.S_idx, self.data[self.S_idx], "vy")

        plt.plot(self.data, "r")

        plt.legend(['Q','R','S','ECG'],loc = 'upper right')
        plt.grid()
        
    def sample_plot(self, s_idx, e_idx): # plot sliced data
        data = self.data.loc[s_idx:e_idx]
        plt.figure(figsize=(18,4))
        plt.plot(self.Q_idx, data[self.Q_idx], "xg")
        plt.plot(self.R_idx, data[self.R_idx], "ob")
        plt.plot(self.S_idx, data[self.S_idx], "vy")

        plt.plot(data, "r")

        plt.legend(['Q','R','S','ECG'],loc = 'upper right')
        plt.grid()
```
