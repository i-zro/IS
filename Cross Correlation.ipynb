```
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
xa = np.linspace(0,3600,361) # 0~3600중 361개 선택, 즉 10개씩
xa
```

```
xb = math.pi * xa / 180
xb
```

```
y1 = np.sin(xb)
y1
```

```
y2 = np.sin(xb - math.pi)
y2
```

```
plt.figure(figsize=(10,5))
plt.grid()
plt.plot(xa, y1, c='k')
plt.plot(xa, y2, linestyle='--', c='r')
```

```
corr1 = np.corrcoef(y1, y1) # corr1은 자신과 자신의 상관관계 = 1
corr2 = np.corrcoef(y1, y2) # corr2는 자신과 반대의 상관관계 = -1
```

```
print(corr1,"\n\n",corr2)
```
